### Define Column Function

We can use in-built functions or define our own functions that we can apply on the DataFrame.

````python
def hundred_plus(col):
  if sum(col) > 100:
    return "Greater than 100"
  return "Not greater than 100"
````

````python
df.apply(hundred_plus)
````

````python
df['odd'].apply(hundred_plus)
````

Why are we getting this error?

When we wrote the hundred_plus function, we expected it to receive a column of values as input, as indicated by the input argument col.

However, when you use `df['odd'].apply(hundred_plus)`, you are trying to apply the hundred_plus function to only one column of the dataframe, specifically the 'odd' column. In this case, the input to the function hundred_plus is a Pandas Series object, which is essentially a one-dimensional array of values, not a column of a dataframe.

Because the function hundred_plus is written to expect a column of values as input (a Pandas Series object), it will not work correctly when applied to a single column of a dataframe. To make the function work in this case, you need to modify it to accept a single value as input and then apply it to each value in the column using the apply method. For example:

```
def hundred_plus(val):
    if val > 100:
        return "Greater than 100"
    return "Not greater than 100"

df['even'] = df['even'].apply(hundred_plus)
df['odd'] = df['odd'].apply(hundred_plus)
```

This will work fine with `df['odd'].apply(hundred_plus)`.

````python
df.apply(hundred_plus, axis=1)
````

### Expanding Results

We can also define functions which return iterable results. In this case, we'll have a function that will return a list with two values. We call this function on a row and we set the values in our return list based on the values in each column of the row. 

Here, we look in the even column and here, we look at the odd column. Now, if we apply this to our DataFrame, we can see it returns an object that has the list of values as the value for each row. Now, there are times where we want to design a function like this, where it'll take those values from the iterable re-return and expand them out so that each value has its own column. We do this using the result type equals expand argument here. When we call it, we can see that the result is a DataFrame with two columns, one for each row value that was returned by our function. Now, as well as applying functions to a whole DataFrame, we can apply functions to individual columns. We're going to want to do this if we just want to base something and the value in a single column, not on summing all the values in a column, or going across a row. Here, we've defined a function. We can see it takes an argument named row, but it treats an argument as an individual value, not as something that it indexes in and grabs different columns, and then it has different return values. Here, what we're doing is we're applying it in the same way we did with the full DataFrame, but we've specified one column to apply it too.

````python
def ret_list(row):
  ret_val = [False, False]
  if row['even'] > 6:
    ret_val[0] = True

  if row['odd'] > 6:
    ret_val[1] = True

  return ret_val
````

````python
df.apply(ret_list, axis=1)
````

````python
df.apply(ret_list, axis=1, result_type ='expand')
````

### Apply to Column

Now, as well as applying functions to a whole DataFrame, we can apply functions to individual columns. We're going to want to do this if we just want to base something and the value in a single column, not on summing all the values in a column, or going across a row. Here, we've defined a function. We can see it takes an argument named row, but it treats an argument as an individual value, not as something that it indexes in and grabs different columns, and then it has different return values. Here, what we're doing is we're applying it in the same way we did with the full DataFrame, but we've specified one column to apply it too.

````python
def div_three(row):
  if row % 3 == 0:
    return "Divisible by 3"
  return "Not divisible by 3"
````

# Numpy Arrays

Numpy is designed for scientific computing. It is built around an Ndarray object which is a N-dimensional array object. It has highly optimized computation tools, and an ecosystem of scientific and mathematical libraries.

There are a lot of the machine learning libraries that expect an array object as an input. And if they don't expect it as an input, they will in fact turn the input into an array object behind the scenes and then return an array object as a result. 

The ndarray object has some characteristics which separate it from python sequences and pandas data frames. 
* First, when you create an ndarray object, its size is set. You cannot resize it without creating a new object. 
* Ndarray objects contain a single data type, unlike pandas data frames. And they are not limited to two dimensions.
* They can be reshaped, returning a new array object, which is how we control these dimensions.

**Where do you use Pandas DataFrame and where do you use Numpy Array?**

Well, you should use an array if you're using high dimensional data, or any dimension that's not two dimensional of a 'single data type'. And your goal is to do complex numerical calculations on this data. 

On the other hand, a Pandas DataFrame is the right choice if you have two dimensional data. Especially with multiple data types and you wish to do some form of data analysis or data visualization on it.

# Choosing a Framework

So, we've looked at a number of different options as alternatives to pandas DataFrames. The question is, how do you choose the right framework for your task? So, at a high level you may want to know whether you want something in the nature of pandas and NumPy which are easy to set up and very immediate to work with and very interactive, or something like PySpark and Dask. 

If this is your choice, it really comes down to the **size of your data**. If you're dealing with big data or if you plan to be dealing with big data, then you should probably go with something like PySpark or Dask. If your data is not big enough to have performance impacts on Pandas or NumPy, it's much easier just to work directly with those.

If you're choosing between Pandas or NumPy, it's really the **nature of your data**. Are you dealing with data that's all of one data types, a numerical data that you need to do highly efficient operations on? Especially operations where the data needs to be represented as a higher dimension matrix then NumPy is the way to go. On the other hand, if you've got tabular data that has data of different data types, then pandas is probably where you want to be. Especially if you're doing just a simple analysis or trying to work towards some visualization.

Now, when choosing between PySpark and Dask, it's really the **nature of your enterprise**. Spark is a much more mature product with a larger ecosystem of support and documentation. So, if you're in an enterprise that's already using Spark or one that is concerned about the maturity of the product, PySpark might be the way to go. 

On the other hand, Dask is Python native and allows you to use the full capabilities of pandas and NumPy DataFrames and arrays. So, if you're starting an evergreen project and you know you want your pipeline all in Python, Dask is a good option.
