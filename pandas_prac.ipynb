When you want to replace something with None, you need to use the list notation. Otherwise, it just messes things in the dataframe. 

For example, if we want to replace all question marks in the age column in titanic:

Avoid: `titanic["age"].replace("?", None)`

Use This: `titanic["age"].replace(["?"], [None])`

## Working with Types and NA Values

```python
titanic.info()
# We see that most columns are objects. Non-numeric data takes a lot of space.
# Age, fare, boat are supposed to be numeric values.
```

```python
titanic['age'].value_counts()

#We see there are 263 '?' values. Pandas can't figure the data type for this.
```

```python
titanic["age"].astype("float")
```

```python
#We need to replace question marks first

titanic["age"].replace(['?'],[None], inplace=True)
```

```python
titanic.age.value_counts(dropna=False)
```

```python
titanic["age"].astype("float")
```

```python
titanic.info()

#age is still shown as object
```

```python
titanic["age"] = titanic["age"].astype("float")
```

```python
titanic.info()

#We see less non-null values in age column. Earlier '?' was treated as value.
```

```python
#Category data type used for finite list of values (eg. male/female)

titanic["sex"].astype("category")

#Notice the output shows the two categories as well.
#Category is more space efficient compared to objects.
```

```python
# Let's check what happens to memory requirements if we change sex from an
# object data type to a category data type

titanic["sex"] = titanic["sex"].astype("category")
```

```python
titanic.info()

#Memory usage earlier was 143.3 KB. Now it's lower - 134.5 KB.
```

```python
titanic['embarked'].value_counts()
```

```python
titanic['embarked'] = titanic['embarked'].astype("category")
```

```python
titanic.info()
```
